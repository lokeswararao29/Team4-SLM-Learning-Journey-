1. Multiplication Table Generator using Functions

```python
def multiplication_table(num):
    for i in range(1, 11):
        print(num, "x", i, "=", num * i)

number = int(input("Enter a number: "))
multiplication_table(number)
```

2. Simple Data Store using Lists


```python
data = []

item1 = input("Enter first item: ")
item2 = input("Enter second item: ")
item3 = input("Enter third item: ")

data.append(item1)
data.append(item2)
data.append(item3)

print("Stored Data:", data)
```

3. Add Student Names into a List

```python
students = []

name1 = input("Enter first student name: ")
name2 = input("Enter second student name: ")
name3 = input("Enter third student name: ")

students.append(name1)
students.append(name2)
students.append(name3)

print("Student List:", students)
```

4. Display all Student Names using a Loop

```python
students = ["Lokesh", "Rahul", "prasad", "Anil"]

for student in students:
    print(student)
```

5. Create a Function to Add Numbers

```python
def add_numbers(a, b):
    result = a + b
    print("Sum:", result)

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

add_numbers(num1, num2)
```

6. Create a Function to Display Student Details

```python
def student_details(name, age, course):
    print("Student Name:", name)
    print("Student Age:", age)
    print("Student Course:", course)

name = input("Enter student name: ")
age = input("Enter student age: ")
course = input("Enter student course: ")

student_details(name, age, course)
```
